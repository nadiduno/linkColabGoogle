# 9) In python, as class method must have _______________ as a function decorator, and the firts parameter of the method will be a reference to _______________ .

Em python, como método de classe deve ter  _______________ como um decorador de função, e o primeiro parâmetro do método será uma referência a _______________ .

 @classmethod and class

O decorador @classmethod em Python é usado para definir um método de classe. Um método de classe é um método que é chamado na classe em vez de em uma instância da classe. O método de classe recebe a própria classe como primeiro argumento, em vez da instância da classe.

class MyClass:
    @classmethod
    def my_method(cls, arg1, arg2):
        # do something with cls and the arguments

```python
class MyClass:
    x = 0

    def __init__(self, y):
        self.y = y

    @classmethod
    def set_x(cls, x):
        cls.x = x

print(MyClass.set_x)
```

# 13) What will this comprehension provide you?

O que essa compreensão lhe proporcionará?

```python
{x : [y for y in range(1,x) if x % y == 0] for x in range (2,100)}
```

Esta é uma compreensão de dicionário em Python que gera um dicionário com chaves de 2 a 99 e valores que são listas de divisores dessas chaves.

As listas de divisores dessas chaves são as listas de números pelos quais cada chave é divisível sem deixar resto. Por exemplo:
Os divisores de 10 são 1, 2, 5 e 10.
