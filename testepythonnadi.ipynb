# 9) In python, as class method must have _______________ as a function decorator, and the firts parameter of the method will be a reference to _______________ .

Em python, como m√©todo de classe deve ter  _______________ como um decorador de fun√ß√£o, e o primeiro par√¢metro do m√©todo ser√° uma refer√™ncia a _______________ .

 @classmethod and class

O decorador @classmethod em Python √© usado para definir um m√©todo de classe. Um m√©todo de classe √© um m√©todo que √© chamado na classe em vez de em uma inst√¢ncia da classe. O m√©todo de classe recebe a pr√≥pria classe como primeiro argumento, em vez da inst√¢ncia da classe.

class MyClass:
    @classmethod
    def my_method(cls, arg1, arg2):
        # do something with cls and the arguments

```python
class MyClass:
    x = 0

    def __init__(self, y):
        self.y = y

    @classmethod
    def set_x(cls, x):
        cls.x = x

print(MyClass.set_x)
```
